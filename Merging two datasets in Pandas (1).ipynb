```python
import pandas as pd
country_code = pd.read_html('https://www.iban.com/country-codes')
```

```python
country_code = country_code[0]
```

```python
country_code = country_code.rename(columns={'Alpha-2 code':'Country Code'})
```

```python
#import libraries i'll be using
import pandas as pd
import requests
import json
import numpy as np
import pandas as pd

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
url = 'http://api.citybik.es/v2/networks'
response = requests.get(url).json()

df = pd.DataFrame()

for i in range (len(response['networks'])):
    temp = pd.DataFrame(
        {
        'Name': [response['networks'][i]['name']],
        'Country': [response['networks'][i]['location']['country']],
        'City': [response['networks'][i]['location']['city']],
        'Latitude': [response['networks'][i]['location']['latitude']],
        'Longitude':[response['networks'][i]['location']['longitude']]
        }
    )
    
    df =pd.concat([df,temp], ignore_index= True)
df.head()
```

```python
df = df.rename(columns={'Country':'Country Code'})
```

```python
df.duplicated().sum()
```

```python
merged_data = df.merge(country_code, left_on='Country Code', right_on='Country Code')
merged_data
```
